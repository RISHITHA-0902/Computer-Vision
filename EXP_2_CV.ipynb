```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read the image
img = cv2.imread("image1.jpg")

# Apply Gaussian Blur
blur = cv2.GaussianBlur(img, (5, 5), 0)

# Resize images to same size (optional but recommended)
blur = cv2.resize(blur, (img.shape[1], img.shape[0]))

# Combine images side by side
combined = np.hstack((img, blur))

# Display
cv2_imshow(combined)

# Save output
cv2.imwrite("blurred_image.jpg", combined)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
