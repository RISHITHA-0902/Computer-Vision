```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read original color image
img = cv2.imread("image1.jpg")

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Create kernel
kernel = np.ones((5, 5), np.uint8)

# Apply dilation
dilated = cv2.dilate(gray, kernel, iterations=1)

# Convert dilated image to BGR
dilated_bgr = cv2.cvtColor(dilated, cv2.COLOR_GRAY2BGR)

# Combine images side by side
combined = np.hstack((img, dilated_bgr))

# Display using cv2_imshow
cv2_imshow(combined)

# Save
cv2.imwrite("dilated_image.jpg", combined)

# cv2.waitKey(0) # Not needed with cv2_imshow in Colab
# cv2.destroyAllWindows() # Not needed with cv2_imshow in Colab
```
